```python
import pandas as pd
import numpy as np
import matplotlib as plot
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
dataset
```

```python
# drop the "user-id" column
dataset=dataset.drop("User ID",axis=1)
dataset
```

```python
dataset.columns
```

```python
#split the input and output data
independent=dataset[['Age', 'EstimatedSalary','Gender_Male']]
```

```python
dependent=dataset[['Purchased']]
```

```python
# # Split the data for train and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.3, random_state=0)
```

```python
# scale the data usig standard scaler
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
x_train=SS.fit_transform(x_train)
x_test=SS.transform(x_test)
```

```python
#Create a model using CV grid search 
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

param_grid= {'criterion': ['gini', 'entropy', 'log_loss'],
             'splitter':['best', 'random'],
             'max_features':['sqrt', 'log2'],
             'splitter': ['best', 'random']
            }
grid=GridSearchCV(DecisionTreeClassifier(),param_grid,cv=5,refit=True,verbose=3,scoring='f1_weighted')
grid.fit(x_train,y_train)
```

```python
best_param=grid.best_params_
print ("Best paramenter: {}".format(best_param))
```

```python
cv_results=grid.cv_results_
table=pd.DataFrame.from_dict(cv_results)
table
```

```python
#finding roc_AUC_score value
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
