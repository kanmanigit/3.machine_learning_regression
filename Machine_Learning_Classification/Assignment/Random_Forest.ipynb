```python
import pandas as pd
import numpy as np
import matplotlib as plot
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset, drop_first=True, dtype=int)
dataset
```

```python
dataset.columns
```

```python
#Split the data
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset[['classification_yes']]
```

```python
dataset['classification_yes'].value_counts()
```

```python
#Split the datas for training and test data set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30, random_state=0)
#x_train,x_test
```

```python
#scale the data to a range using Standard scaler
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
x_train=SS.fit_transform(x_train)
x_test=SS.transform(x_test)
#x_train,x_test
```

```python
#Model creation 
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

param_grid={'n_estimators':[10],
            'criterion':['entropy','gini','log_loss'],
            'max_features':['sqrt','log2', None],
            'random_state':[0]
            }
model=GridSearchCV(RandomForestClassifier(), param_grid, n_jobs=-1,scoring='f1_weighted',refit=True,cv=10,verbose=3)

model.fit(x_train,y_train)
ypred=model.predict(x_test)
```

```python
# finding the best parameters
print("Best parameters {}".format(model.best_params_))
```

```python
#finding the CV results
cv_results=model.cv_results_
table=pd.DataFrame.from_dict(cv_results)
table
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,ypred)
print (cm)
```

```python
#classification report
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,ypred)
print(clf_report)
```
