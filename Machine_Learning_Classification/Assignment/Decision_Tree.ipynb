```python
import pandas as pd
import numpy as np
import matplotlib as plot
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset.columns
```

```python
dataset=pd.get_dummies(dataset, drop_first=True, dtype=int)
dataset
```

```python
dataset.columns
```

```python
#Split the data
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset[['classification_yes']]
```

```python
#independent,dependent
```

```python
#Split the datas for training and test data set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30, random_state=0)
#x_train,x_test
```

```python
#scale the data to a range using Standard scaler
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
x_train=SS.fit_transform(x_train)
x_test=SS.transform(x_test)
#x_train,x_test
```

```python
#Model creation using CV search and linear model
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

param_grid={'criterion': ['gini', 'entropy', 'log_loss'],
             'splitter':['best', 'random'],
             'max_features':['sqrt', 'log2']
           }
model=GridSearchCV(DecisionTreeClassifier(), param_grid, refit=True,scoring='f1_weighted',cv=5)
model.fit(x_train,y_train)
                       
```

```python
#Find the best parameter
print("Best parameter: {}".format(model.best_params_))
```

```python
#Print the CV results in table
cv_result=model.cv_results_
table=pd.DataFrame.from_dict(cv_result)
print(table)
```

```python
#confusion matrix

y_pred=model.predict(x_test)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
# classification report
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_pred)
print(clf_report)
```

```python
#finding ROC_AUC score
from sklearn.metrics import roc_auc_score
roc_auc_score=roc_auc_score(y_test,model.predict_proba(x_test)[:,1])
print(roc_auc_score)
```
