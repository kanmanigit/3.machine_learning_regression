```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
plot.scatter(dataset["Annual Income (k$)"],dataset["Spending Score (1-100)"])
plot.xlabel("Annual Income (k$)")
plot.ylabel("Spending Score (1-100)")
plot.title("customer data")
```

```python
#Subset of dataset....
x=dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import AffinityPropagation
af=AffinityPropagation(random_state=5)
af.fit_predict(x)
```

```python
n_clusters=af.cluster_centers_indices_

print("No of clusters is: ",len(n_clusters))
```

```python
af.convergence_iter
```

```python
label=af.labels_
label
```

```python
#creating table with the exisiting data and append no of clusters with that
supervised=pd.DataFrame(dataset)
supervised["Cluster_Group"]=label

```

```python
supervised.to_csv("AP_Clustering.csv",index=False)
```

```python
sns.lmplot(supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],fit_reg=False)
```
